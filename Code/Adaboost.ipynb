```python
names = ['id', 'age', 'exp', 'inc', 'zip', 'family', 
         'ccavg', 'edu', 'mortgage', 'loan', 
         'securities', 'cd', 'online', 'cc']
```

#### dropping columns

```python
data.drop(columns=['id','zip','exp'], axis=1, inplace=True)
```

```python
cat_Attr = ["family", "edu", "securities", "cd", "online", "cc", "loan"]
data[cat_Attr] = data[cat_Attr].astype('category')
```

```python
cat_Attr2 = ["family", "edu", "securities", "cd", "online", "cc"]
```

#### Creating dummy variables

```python
data = pd.get_dummies(data, columns=cat_Attr2)
```

```python
data.columns
```

```python
data2 = data.drop('loan',axis=1)
```

```python
names = list(data2.columns.values)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max = MinMaxScaler()
```

```python
std_data = pd.DataFrame(min_max.fit_transform(data2), columns = names)
```

```python
std_data.describe()
```

```python
bank = pd.concat([data2, data['loan']], axis=1)
```

```python
bank.head()
```

#### Split the data

```python
X, y = bank.iloc[:,:-1], bank.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

#### Model building

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier() 
clf = AdaBoostClassifier(n_estimators=100, base_estimator=dt,learning_rate=1)
```

```python
clf.fit(X_train, y_train)
```

```python
pred = clf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
print(confusion_matrix(y_test, pred))
print("\n")
print(classification_report(y_test, pred))
print("\n")
print("Accuracy = "+str(accuracy_score(y_test, pred)))
```
